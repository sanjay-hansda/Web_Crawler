```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import time
import random

# Base URL of the website
base_url = "https://thebetterindia.com/topics/sustainability/page/"

# Number of pages to scrape (e.g., 45 pages)
total_pages = 45

# Directory to save HTML files
html_dir = "html_files_1"
os.makedirs(html_dir, exist_ok=True)

# List of proxies (replace with your list)
proxies = [
    "20.27.86.185:8080",
    "67.43.236.20:11859",
    "67.43.228.251:14791",
    "77.93.143.38:26654",
    "138.68.235.51:80",
    "188.40.59.208:3128",
    "47.238.67.96:8888",
    "81.169.213.169:8888",
    "46.47.197.210:3128",
    "45.140.143.77:18080",
    "51.68.175.56:1080",
    "158.255.77.169:80",
    "71.14.218.2:8080",
    "5.106.6.235:80",
    "94.70.195.145:8080",
    "143.42.66.91:80",
    "198.49.68.80:80",
    "47.251.43.115:33333",
    "162.223.90.130:80",
    "67.43.227.226:11305",
    "34.81.160.132:80",
    "85.215.64.49:80",
    "50.223.246.237:80",
    "50.174.7.159:80",
    "50.207.199.87:80",
    "41.207.187.178:80",
    "32.223.6.94:80",
    "82.119.96.254:80",
    "50.207.199.80:80",
    "50.207.199.83:80",
    "50.174.7.153:80",
    "50.169.37.50:80",
    "50.232.104.86:80",
    "50.175.212.66:80",
    "50.217.226.47:80",
    "50.239.72.16:80",
    "50.217.226.40:80",
    "50.221.74.130:80",
    "190.58.248.86:80",
    "50.175.212.74:80",
    "50.207.199.82:80",
    "50.174.7.152:80",
    "50.122.86.118:80",
    "66.191.31.158:80",
    "65.108.195.47:8080",
    "8.210.232.181:7888",
    "41.74.91.244:80",
    "3.136.29.104:80",
    "200.174.198.86:8888",
    "13.208.56.180:80",
    "35.72.118.126:80",
    "43.202.154.212:80",
    "3.37.125.76:3128",
    "35.76.62.196:80",
    "18.228.149.161:80",
    "46.51.249.135:3128",
    "3.139.242.184:80",
    "43.201.121.81:80",
    "54.233.119.172:3128",
    "52.196.1.182:80",
    "3.12.144.146:3128",
    "18.228.198.164:80",
    "52.67.10.183:80",
    "54.37.214.253:8080",
    "204.236.176.61:3128",
    "3.130.65.162:3128",
    "51.16.199.206:3128",
    "13.246.209.48:1080",
    "54.179.39.14:3128",
    "13.246.184.110:3128",
    "51.17.58.162:3128",
    "52.65.193.254:3128",
    "3.97.167.115:3128",
    "51.16.179.113:1080",
    "63.32.1.88:3128",
    "54.179.44.51:3128",
    "13.55.210.141:3128",
    "54.152.3.36:80",
    "13.59.156.167:3128",
    "45.61.159.42:3128",
    "34.244.90.35:80",
    "47.57.100.251:8099",
    "158.255.77.168:80",
    "54.248.238.110:80",
    "50.169.222.243:80",
    "200.250.131.218:80",
    "50.169.222.241:80",
    "141.11.103.136:8080",
    "158.255.77.166:80",
    "47.56.110.204:8989",
    "50.217.226.41:80",
    "50.207.199.86:80",
    "0.0.0.0:80",
    "87.248.129.26:80",
    "211.128.96.206:80",
    "68.185.57.66:80",
    "50.231.104.58:80",
    "50.174.7.156:80",
    "50.207.199.81:80",
    "127.0.0.7:80",
    "103.154.230.190:8080",
    "103.22.99.12:2020",
    "103.53.78.123:8050",
    "103.166.159.163:8080",
    "47.251.122.81:8888",
    "137.184.100.135:80",
    "13.37.89.201:80",
    "13.37.59.99:3128",
    "13.36.104.85:80",
    "13.36.113.81:3128",
    "13.36.87.105:3128",
    "15.236.106.236:3128",
    "44.218.183.55:80",
    "44.195.247.145:80",
    "50.202.75.26:80",
    "165.232.129.150:80",
    "50.239.72.18:80",
    "47.245.34.6:3389",
    "54.67.125.45:3128",
    "13.56.192.187:80",
    "3.96.208.91:3128",
    "23.247.136.248:80",
    "35.79.120.242:3128",
    "43.200.77.128:3128",
    "43.129.201.43:443",
    "13.48.109.48:3128",
    "52.63.129.110:3128",
    "99.80.11.54:3128",
    "51.20.50.149:3128",
    "15.156.24.206:3128",
    "13.213.114.238:3128",
    "3.97.176.251:3128",
    "16.16.239.39:3128",
    "54.228.164.102:3128",
    "51.20.19.159:3128",
    "3.212.148.199:3128",
    "103.237.144.232:1311",
    "198.74.51.79:8888",
    "23.88.116.40:80",
    "31.40.248.2:8080",
    "216.229.112.25:8080",
    "45.174.95.137:8080",
    "147.45.49.234:3128",
    "188.132.222.230:8080",
    "181.233.62.9:999",
    "187.102.236.209:999",
    "182.93.82.8:8080",
    "185.82.99.210:8080",
    "103.13.193.247:8080",
    "103.169.132.70:8085",
    "72.10.160.172:11723",
    "8.210.38.209:1081",
    "195.114.209.50:80",
    "91.92.96.210:8080",
    "123.30.154.171:7777",
    "8.209.210.171:3389",
    "43.156.59.228:80",
    "213.134.208.187:8080",
    "47.88.59.79:82",
    "51.254.78.223:80",
    "159.65.245.255:80",
    "47.237.13.153:10086",
    "66.29.154.103:3128",
    "3.71.239.218:3128",
    "143.42.191.48:80",
    "31.47.58.37:80",
    "143.198.226.25:80",
    "50.239.72.19:80",
    "184.169.154.119:80",
    "3.126.147.182:80",
    "3.122.84.99:3128",
    "3.124.133.93:3128",
    "18.185.169.150:3128",
    "3.78.92.159:3128",
    "116.125.141.115:80",
    "23.94.136.205:80",
    "18.175.228.65:41451",
    "34.238.112.235:8082",
    "204.236.137.68:80",
    "52.73.224.54:3128",
    "44.219.175.186:80",
    "202.146.228.254:8088",
    "179.228.241.22:8080",
    "217.24.162.52:8080",
    "38.183.146.97:8090",
    "116.203.139.209:5678",
    "202.61.199.166:80",
    "219.65.73.81:80",
    "185.212.60.63:80",
    "47.252.11.233:9080",
    "47.74.46.81:31433",
    "104.238.160.36:80",
    "3.127.62.252:80",
    "3.129.184.210:80",
    "51.75.206.209:80",
    "84.39.112.144:3128",
    "97.74.87.226:80",
    "185.212.60.62:80",
    "3.123.150.192:80",
    "45.151.70.164:8080",
    "190.94.213.5:999",
    "103.48.68.76:83",
    "12.88.29.66:9080",
    "13.37.73.214:80",
    "47.245.117.43:80",
    "52.16.232.164:3128",
    "3.90.100.12:80",
    "8.219.97.248:80",
    "203.115.101.53:82",
    "23.82.137.157:80",
    "190.103.177.131:80",
    "149.202.91.219:80",
    "72.10.160.174:8005",
    "72.10.160.91:13219",
    "23.82.137.156:80",
    "72.10.160.170:5475",
    "146.59.202.70:80",
    "103.134.222.186:8080",
    "145.239.196.123:80",
    "142.44.210.174:80",
    "41.59.90.171:80",
    "66.29.154.105:3128",
    "222.252.194.204:8080",
    "147.182.180.242:80",
    "8.211.194.85:4444",
    "38.54.116.9:8118",
    "8.211.49.86:8443",
    "3.21.101.158:3128",
    "113.160.133.32:8080",
    "63.35.64.177:3128",
    "113.23.98.81:8080",
    "203.192.199.158:53",
    "154.193.8.70:808",
    "45.167.124.141:999",
    "112.201.237.199:8082",
    "202.62.75.17:83",
    "209.97.150.167:8080",
    "27.79.177.126:16000",
    "3.141.217.225:80",
    "192.73.244.36:80",
    "144.126.216.57:80",
    "209.45.108.230:1011",
    "34.81.72.31:80",
    "13.38.176.104:3128",
    "3.108.115.48:3128",
    "38.137.252.10:999",
    "179.61.98.3:999",
    "154.193.38.121:808",
    "45.233.90.10:443",
    "185.206.83.246:8080",
    "36.91.220.131:8080",
    "1.1.220.100:8080",
    "113.23.157.39:1231",
    "190.110.34.108:999",
    "103.162.62.250:8181",
    "181.45.78.113:8080",
    "202.57.25.196:1111",
    "45.159.113.153:3128",
    "115.127.28.10:8674",
    "103.84.176.101:8083",
    "183.88.224.86:8080",
    "23.82.137.161:80",
    "63.143.57.117:80",
    "47.251.73.54:4145",
    "18.223.25.15:80",
    "149.129.255.179:9050",
    "23.82.137.158:80",
    "13.38.153.36:80",
    "3.127.121.101:80",
    "45.144.64.153:8080",
    "180.180.123.40:8080",
    "35.154.71.72:3128",
    "87.248.129.32:80",
    "203.115.101.55:82",
    "43.224.116.125:8080",
    "200.39.139.65:999",
    "103.171.255.59:8080",
    "135.225.231.41:8080",
    "92.51.126.42:8080",
    "159.203.61.169:3128",
    "195.26.249.42:3128",
    "139.162.78.109:8080",
    "8.221.141.88:8083",
    "104.225.220.233:80",
    "8.217.124.178:49440",
    "202.169.51.46:8080",
    "38.183.146.170:8090",
    "113.160.184.87:8080",
    "103.163.244.105:83",
    "200.116.230.212:8080",
    "175.100.70.72:8080",
    "89.117.130.19:80",
    "23.82.137.162:80",
    "175.139.233.78:80",
    "93.127.163.52:80",
    "78.28.152.113:80",
    "72.10.160.171:16399",
    "8.215.105.136:7777",
    "207.180.254.198:80",
    "93.113.63.73:33100"
]

# Headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}

# Function to fetch HTML content from a page using a proxy
def fetch_html(page_number):
    url = f"{base_url}{page_number}/"
    for proxy in proxies:
        try:
            proxy_dict = {"http": f"http://{proxy}", "https": f"http://{proxy}"}
            response = requests.get(url, headers=headers, proxies=proxy_dict, timeout=10)
            if response.status_code == 200:
                return response.text
            else:
                print(f"Failed to fetch page {page_number} with proxy {proxy}. Status code: {response.status_code}")
        except Exception as e:
            print(f"Error with proxy {proxy}: {e}")
    print(f"All proxies failed for page {page_number}.")
    return None

# Function to save HTML content to a file
def save_html(html_content, page_number):
    file_path = os.path.join(html_dir, f"page_{page_number}.txt")
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(html_content)
    print(f"Saved HTML for page {page_number}")

# Function to extract data from HTML content
def extract_data(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    titles = []
    links = []
    dates = []

    article_cards = soup.find_all('div', class_='article-card')
    for card in article_cards:
        title = card.find('h3', class_='heading').text.strip()
        link = card.find('a')['href']
        date = card.find('time')['datetime']
        titles.append(title)
        links.append(link)
        dates.append(date)

    return titles, links, dates

# Main function to scrape all pages
def scrape_all_pages():
    all_titles = []
    all_links = []
    all_dates = []

    for page_number in range(1, total_pages + 1):
        print(f"Fetching page {page_number}...")
        html_content = fetch_html(page_number)
        if html_content:
            save_html(html_content, page_number)
            titles, links, dates = extract_data(html_content)
            all_titles.extend(titles)
            all_links.extend(links)
            all_dates.extend(dates)

        # Add a delay between requests to avoid being blocked
        time.sleep(random.uniform(2, 5))  # Random delay between 2 and 5 seconds

    # Save all data to an Excel file
    data = {
        'Title': all_titles,
        'Link': all_links,
        'Publish Date': all_dates
    }
    df = pd.DataFrame(data)
    df.to_excel('all_articles.xlsx', index=False)
    print("All data extracted and saved to 'all_articles.xlsx'")

# Run the scraper
scrape_all_pages()
```
